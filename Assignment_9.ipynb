3.What do the functions os.getcwd() and os.chdir() do?

Ans : os.getcwd() method tells us the location of current woking directory (CWD). whereas os.chdir() method in Python used to change the current working directory to specified path. These functions are similar to linux commands pwd and cd

```python
import os 
print(os.getcwd())
path = r'C:\Users\samir saiyed\Documents\final submission'
os.chdir(path)
print(os.getcwd())
```

6.What are the three “mode” arguments that can be passed to the open() function?

Ans : A file can be Accessed in python using open() function. open function takes two arguments filename and mode of operation(optional). if mode id not provided the default mode of opening is read mode                                       So, the syntax being : open(filename , mode)

- 'r' : Read mode : This is the default mode for open().the file id opened and a pointer is positioned at the beginning of the files content.                                                                                                         
- 'w' : Write mode : Using this mode will overwrite any existing content in a file. if the given file does not exist, a new one will be created.                                                                                                       
- 'r+' : Read,Write mode : Use this mode if you need to simulataneously read and write to a file.                          
- 'a' : Append mode : with this mode the user can append the data without overwriting any already existing data in the file.     
- 'a+' - Append and Read mode : with this mode the user can append the data without overwriting any already existing data in the file.
- 'x' : Exclusive creating mode : this mode is for the sole purpose of creating new files.use this mode if you know the file to be written doesn't exist beforehand.

8.How do you tell the difference between read() and readlines()?

Ans : The main diffrence is that read() will read the whole file at once and print out the first characters that take up as many bytes as you specify in the parenthesis                                                                               

- Whereas the readline() that will read and print out only the first characters that take up as many bytes as you specify in the parenthasis. you may want to use readline() when you're reading files that are too big for your RAM.
- The read() would treat each character in the file separately, meaning that the iteration would happen for every character.
- The readline() function, on the other hand , only a single line of the file. this means that if the first line of the file were three lines long, the readline() function would only parse (or iterte/operate) on the first line of the file.
